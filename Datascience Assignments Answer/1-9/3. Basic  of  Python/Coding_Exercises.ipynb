### Exercise 1: Prime Numbers

```python
def is_prime(number):
    """Check if a number is a prime number."""
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

# Test the function
num = 29
print(f"Is {num} a prime number? {is_prime(num)}")
```

### Exercise 4: Word Counter

```python
from collections import Counter

def word_counter(text):
    """Counts occurrences of each word in the input text."""
    words = text.split()
    counts = Counter(words)
    return counts

# Test input
input_text = "This is a sample text. This text will be used to demonstrate the word counter."

# Count words
word_counts = word_counter(input_text)

# Display output
for word, count in word_counts.items():
    print(f"'{word}': {count}")
```

### Exercise 5: Check for Palindrome

```python
import string

def is_palindrome(text):
    """Checks whether the given text is a palindrome, ignoring punctuation, case, and spaces."""
    cleaned = ''.join(char.lower() for char in text if char.isalnum())
    return cleaned == cleaned[::-1]

# Test input
sample_text = "Racecar"
print(f"Is '{sample_text}' a palindrome? {is_palindrome(sample_text)}")
```
