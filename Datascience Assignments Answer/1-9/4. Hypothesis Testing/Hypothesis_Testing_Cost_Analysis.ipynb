# Hypothesis Testing – Weekly Operating Cost Analysis

**Background**: Bombay Hospitality Ltd. uses the cost model `W = 1000 + 5X` to estimate weekly franchise operating costs. Restaurant owners believe the cost has increased.

## 1. Hypotheses
- Null Hypothesis (H₀): The weekly operating cost has not increased; it follows the model `W = 1000 + 5X`
- Alternative Hypothesis (H₁): The weekly operating cost **has increased** (one-tailed test)

```python
# Given values
sample_mean = 3050  # Rs.
x_mean = 600  # units produced per week
n = 25  # sample size
std_dev_units = 25  # standard deviation of units produced

# Calculate theoretical mean
mu = 1000 + 5 * x_mean

# Calculate standard deviation of cost
sigma = 5 * std_dev_units

# Calculate test statistic (Z)
z = (sample_mean - mu) / (sigma / (n ** 0.5))
z
```

## 2. Calculate the Critical Value
Using a one-tailed test at α = 0.05 significance level

```python
from scipy.stats import norm

alpha = 0.05
critical_value = norm.ppf(1 - alpha)
critical_value
```

## 3. Make a Decision

```python
if z > critical_value:
    print("Reject the null hypothesis – Weekly operating cost has increased.")
else:
    print("Fail to reject the null hypothesis – No sufficient evidence of increase.")
```

## 4. Conclusion
This test allows us to decide whether the observed costs are statistically significantly higher than predicted by the model.
